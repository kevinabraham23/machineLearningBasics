```python
df['Data']= df['Data'].apply(lambda x: x.split()[0])
df.head()
```

```python
#df['Month'] = df['Data'].apply(lambda x:x.split('/')[0]) makes it into an  object
df['Month'] = pd.to_datetime(df['Data']).dt.month
df['Day'] = pd.to_datetime(df['Time']).dt.day
df['Hour'] = pd.to_datetime(df['Time']).dt.hour
df['Minute'] = pd.to_datetime(df['Time']).dt.minute
df['Second'] = pd.to_datetime(df['Time']).dt.second
```

```python
df
```

```python
df.dtypes
```

```python
import re
```

```python
df['riseHour'] = df['TimeSunRise'].apply(lambda x: re.search(r'^\d+',x).group(0)).astype(int)
df['riseMinute'] = df['TimeSunRise'].apply(lambda x: re.search(r'(?<=\:)\d+(?=\:)',x).group(0)).astype(int)

df['setHour'] = df['TimeSunSet'].apply(lambda x: re.search(r'^\d+',x).group(0)).astype(int)
df['setMinute'] = df['TimeSunSet'].apply(lambda x: re.search(r'(?<=\:)\d+(?=\:)',x).group(0)).astype(int)
```

```python
df.drop(['UNIXTime','Data','Time','TimeSunRise','TimeSunSet'],axis=1,inplace=True)
```

### Select K Best Method

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
bestfeatures = SelectKBest(score_func=chi2)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
Y_train=label_encoder.fit_transform(target)
```

```python
target_cont = df['Radiation'].apply(lambda x : int(x*100))
```

```python
from sklearn.preprocessing import MinMaxScaler

scaled_input_features = MinMaxScaler().fit_transform(input_features)
```

```python
fit  = bestfeatures.fit(scaled_input_features,Y_train)
```

```python
scores = pd.DataFrame(fit.scores_)
columns = pd.DataFrame(input_features.columns)
```

```python
featureScores = pd.concat([columns,scores],axis=1)
```

```python
featureScores.columns = ['Features','Scores']
featureScores
```

```python
featureScores.sort_values(by='Scores',ascending=False,inplace=True)
```

```python
featureScores
```

```python
plt.bar(featureScores.Features,featureScores.Scores)
plt.xticks(rotation=90)
plt.show()
```
